### 📝 Exercise: Build a Nutrition Coach Chatbot
Goal:
Create a Gradio `ChatInterface` that:
- Uses a system prompt to set the chatbot’s persona.
- Accepts a user message.
- Keeps track of conversation history.
- Returns responses in the style of a nutrition expert.

```python
# Imports

import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override= True)

openai_api_key= os.getenv("OPENAI_API_KEY")

if openai_api_key:
    print(f"OpenAI API Key exists and begins with {openai_api_key[:8]}....")
else:
    print("OpenAI API Key is not set")

openai_client = OpenAI()
MODEL = "gpt-4.1-mini"
```

```python
# Create a system Prompt

system_prompt = """
You are a certifie nutrition coach with an experience of more than 20+ years.
Give clear, practical advice about healthy eating, diet planning, and lifestyle
improvements. Whenever user ask a question related to your field you should 
respond like - 'As your nutrition coach, here's my advice' If user asks other 
questions which are not related to the niche then no need to reply with 
'As your nutrition coach, here's my advice'. 
"""
```

```python
def chat(message, history):

    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]

    stream = openai_client.chat.completions.create(
        model= MODEL,
        messages= messages,
        stream= True
    )

    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn= chat, title= "Nutrition Coach Chatbot").launch()
```
